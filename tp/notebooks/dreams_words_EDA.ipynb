```python
import pandas as pd

def p1(x):
    return x.quantile(0.01)
def p5(x):
    return x.quantile(0.05)
def p10(x):
    return x.quantile(0.10)
def p25(x):
    return x.quantile(0.25)
def p50(x):
    return x.quantile(0.50)
def p75(x):
    return x.quantile(0.75)
def p90(x):
    return x.quantile(0.90)
def p95(x):
    return x.quantile(0.95)
def p99(x):
    return x.quantile(0.99)
```

```python
dreams_clean=pd.read_csv("../data/dreams_clean.csv",delimiter=";",encoding="latin-1")
```

```python
dreams_clean["group"].unique()
```

## Grupos de sueÃ±os pre-seleccionados

```python
group_list=[
"Chuck: a physical scientist",
"Kenneth",
"Melvin (Melora's husband)",
"Miami Home-Lab: Home",
"Miami Home-Lab: Lab",
"The Natural Scientist",
"Norman: a child molester",
"Pegasus: a factory worker",
"Phil 2: late 20s",
"Phil 3: retirement",
"The Physiologist",
"Robert Bosnak: A dream analyst",
'Vietnam Vet: 1970-2008 war dreams',
'Vietnam Vet: 2015 dreams', 
'Vietnam Vet: 2016-17 dreams']
```

```python
dreams_groups=dreams_clean[dreams_clean["group"].isin(group_list)]
```

## Cantidad de sueÃ±os y palabras sin ningun filtro

```python
words=dreams_groups[["group","words"]].groupby(['group']).agg({'group':['size'],'words':['mean',p1,p5,p10,p25,p50,p75,p90,p95,p99]})
words.columns = ["_".join(x) for x in words.columns.ravel()]
print('\033[1m'+"CANTIDAD TOTAL DE SUEÃ‘OS: ",words.iloc[:,0].sum())
words
```

## Si filtramos con 50 palabras

```python
dreams_groups_n50=dreams_groups[dreams_groups["words"]>50]
words_50=dreams_groups_n50[["group","words"]].groupby(['group']).agg({'group':['size'],'words':['mean',p1,p5,p10,p25,p50,p75,p90,p95,p99]})
words_50.columns = ["_".join(x) for x in words_50.columns.ravel()]
print('\033[1m'+"CANTIDAD TOTAL DE SUEÃ‘OS CON +50 PALABRAS: ",words_50.iloc[:,0].sum())
words_50
```

## Si filtramos con 75 palabras

```python
dreams_groups_n75=dreams_groups[dreams_groups["words"]>75]
words_75=dreams_groups_n75[["group","words"]].groupby(['group']).agg({'group':['size'],'words':['mean',p1,p5,p10,p25,p50,p75,p90,p95,p99]})
words_50.columns = ["_".join(x) for x in words_50.columns.ravel()]
print('\033[1m'+"CANTIDAD TOTAL DE SUEÃ‘OS CON +75 PALABRAS: ",words_75.iloc[:,0].sum())
words_75
```
